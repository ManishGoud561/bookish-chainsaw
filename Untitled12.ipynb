```python
n = 5
a, b = 0, 1
for i in range(n):
    print(a, end=" ")
    a, b = b, a+b
print()
```

```python
n = int(input("Enter a num:"))
a, b = 0, 1
for i in range(n):
    print(a, end=" ")
    a, b = b, a+b
print()
```

```python
with open("single.txt") as f:
    n = int(f.readline())

a, b = 0, 1
for i in range(n):
    print(a, end=" ")
    a, b = b, a+b
print()
```

```python
with open("multiple.txt") as f:
    for line in f:
        n = int(line.split()[0])
        a, b = 0, 1
        for i in range(n):
            print(a, end=" ")
            a, b = b, a+b
        print()
```

program 10

```python
n = 1234
rev = 0
while n > 0:
    rev = rev * 10 + n % 10
    n //= 10
print(rev)
```

```python
n = int(input("Enter a num:"))
rev = 0
while n > 0:
    rev = rev * 10 + n % 10
    n //= 10
print(rev)
```

```python
with open("single.txt") as f:
    n = int(f.readline())

rev = 0
while n > 0:
    rev = rev * 10 + n % 10
    n //= 10
print(rev)
```

```python
with open("multiple.txt") as f:
    for line in f:
        n = int(line.split()[0])
        rev = 0
        while n > 0:
            rev = rev * 10 + n % 10
            n //= 10
        print(rev)
```

program 11

```python
n = 121
temp = n
rev = 0
while n > 0:
    rev = rev * 10 + n % 10
    n //= 10
print("Palindrome" if temp == rev else "Not Palindrome")
```

```python
n = int(input("Enter a num:"))
temp = n
rev = 0
while n > 0:
    rev = rev * 10 + n % 10
    n //= 10
print("Palindrome" if temp == rev else "Not Palindrome")
```

```python
with open("single.txt") as f:
    n = int(f.readline())

temp = n
rev = 0
while n > 0:
    rev = rev * 10 + n % 10
    n //= 10
print("Palindrome" if temp == rev else "Not Palindrome")
```

```python
with open("multiple.txt") as f:
    for line in f:
        n = int(line.split()[0])
        temp = n
        rev = 0
        while n > 0:
            rev = rev * 10 + n % 10
            n //= 10
        print("Palindrome" if temp == rev else "Not Palindrome")
```
