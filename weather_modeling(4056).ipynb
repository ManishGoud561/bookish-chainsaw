```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
time = np.array([0, 4, 8, 12, 16, 20, 24])
temperature = np.array([22, 24, 30, 35, 32, 28, 24])
```

```python
coefficients = np.polyfit(time, temperature, 2)
a, b, c = coefficients
```

```python
t_values = np.arange(0, 25, 1)
predicted_temp = a * t_values**2 + b * t_values + c
```

```python
print("Quadratic Model Coefficients:")
print("a =", a)
print("b =", b)
print("c =", c)
```

```python
plt.plot(time, temperature, 'o', label="Actual Data")
plt.plot(t_values, predicted_temp, label="Predicted Curve")
plt.xlabel("Time (Hours)")
plt.ylabel("Temperature (Â°C)")
plt.title("Weather Prediction Using Quadratic Model")
plt.legend()
plt.show()
```
